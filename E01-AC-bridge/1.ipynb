```python
from math import pi

f = 996
w = 2 * pi * f
```

```python
# 电感串联，~10mH
R2 = 100
C0 = 1e-7
R3 = 988.1
R0 = 34

Lx = (R2 * R3 * C0) / (1 + (w * R0 * C0)**2)
rx = (w**2 * C0**2 * R0 * R2 * R3) / (1 + (w * R0 * C0)**2)
Q = 1 / (w * R0 * C0)
print(Lx * 1e3, "mH")
print(rx)
print(Q)
print(abs(Lx * 1e3 - 10) / 10 * 100, "%")
```

```python
# 电感并联，~1mH
R2 = 100
C0 = 1e-8
R3 = 1001.0
R0 = 9150

Lx = R2 * R3 * C0
rx = (R2 * R3) / (R0)
Q = w * C0 * R0
print(Lx * 1e3, "mH")
print(rx)
print(Q)
```

```python
# 电容串联，~0.1uF
R4 = 1000
C0 = 1e-7
R3 = 6
R0 = 1024.4

Cx = (R4 * C0) / R3
rx = R3 * R0 / R4
D = w * C0 * R0
print(Cx * 1e6, "uF")
print(rx)
print(D)
```
